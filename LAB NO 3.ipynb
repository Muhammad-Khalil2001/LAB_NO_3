```python
import tensorflow as tf
import matplotlib.pyplot as plt

w=tf.Variable([0.2], dtype=tf.float32)
b=tf.Variable([-0.2], dtype=tf.float32)

y=tf.Variable([3, 5, 7, 9], dtype=tf.float32)
x=tf.Variable([2,4,6,8], dtype=tf.float32)

Linear_Model = w * x + b

print("Initial Weight:",w)
print("Initial Bias:",b)

print("Predicted Output Is:")
predicted_output = (Linear_Model)
print(predicted_output)

squared_deltas = tf.square(Linear_Model - y)
loss = tf.reduce_sum(squared_deltas)

print("Loss or Error is:")
print(loss)

plt.scatter ([2, 4, 6, 8], [3, 5, 7, 9], label='Actual Data')
plt.plot([2, 4, 6, 8], predicted_output, color='red', label='Predicted Output')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Data Points and Predicted Output')
plt.legend()
plt.show()
```
